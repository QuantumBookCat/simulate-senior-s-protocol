```python
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister
import numpy as np
import qiskit
%matplotlib inline
import random
```

```python
# Step0 : message||Hash(message) and Pre-shared Key

# message
message = "Hello world"
print("message : {}".format(message))

length_message_alice = len(message)
print("length_message_alice : {}".format(length_message_alice))


# m
def Message_01str(strr,length):
    m0=[]
    m0.clear()
    for i in range(length):
        m0.append(ord(strr[i]))
    print("m0 : {}".format(m0))

    m1=[]
    m1.clear()
    for i in range(length):
        m1.append(np.binary_repr(ord(strr[i]),8))
    print("m1 : {}".format(m1))

    m2=""
    for i in range(length):
        m2+=(np.binary_repr(ord(strr[i]),8))    
    print("m2 : {}".format(m2))

    m=[]
    m.clear()
    for i in range(len(m2)):
        m.append(int(m2[i]))    
    #print("m : {}".format(m))
    
    return m



m_alice=[]
m_alice.clear()
m_alice = Message_01str(message,length_message_alice)
print("m_alice : {}".format(m_alice))
#print("length_m_alice : {}".format(len(m_alice)))


# Hash(m)
def Hash_preshared(list1,length,list2):
    list3=[]
    for i in range(length):
        str3=""
        for j in range(8):
            str3+=f'{list1[8*i+j]}'
        list3.append(str3)
    print("m1 = {}".format(list3))
    
    list5=[]
    for i in range(length):
        list5.append(int(list3[i],2))
    print("m2 = {}".format(list5))
    
    list6=[]
    for i in range(length):
        list6.append(int(list3[i],2)%8)
    print("Hm2 = {}".format(list6))
    
    list7=[]
    for i in range(length):
        list7.append(np.binary_repr(int(list3[i],2)%8,3))
    print("Hm1 = {}".format(list7))
    
    str7=""
    for i in range(length):
        str7+=list7[i]
    print("str_Hm1 = {}".format(str7))
    
    for j in range(len(str7)):
        list2.append(int(str7[j]))
    
      

Hm_alice=[]
Hm_alice.clear()       
        
Hash_preshared(m_alice,length_message_alice,Hm_alice)
print("Hm_alice = {}".format(Hm_alice))


# m||H(m)
m_Hm_list=[]
m_Hm_list.clear()
for i in range(len(m_alice)):
    m_Hm_list.append(m_alice[i])

    
for i in range(len(Hm_alice)):
    m_Hm_list.append(Hm_alice[i])
    
print("m_Hm_list : {}".format(m_Hm_list))
print("length_m_Hm_list : {}".format(len(m_Hm_list)))

# N
n = len(m_Hm_list)
print("N : {}".format(n))

# Pre-shared Key
Alice_pre_key=[]
Bob_pre_key=[]
Alice_pre_key.clear()
Bob_pre_key.clear()
for i in range(n):
    num=random.randint(0,1)
    Alice_pre_key.append(num)
    Bob_pre_key.append(num)
    
print("Alice_pre_key : {}".format(Alice_pre_key))
print("Bob_pre_key : {}".format(Bob_pre_key))
```

```python
# Step1 : generate N random single qubits
def random_01str(list,num):
    for i in range(num):
        list.append(random.randint(0,1))
        
#Alice
Nqubits_01=[]
Nqubits_01.clear()
random_01str(Nqubits_01,n)
print("Nqubits_01 ,0(0) or 1(1) : {}".format(Nqubits_01))

Nqubits_ZX=[]
Nqubits_ZX.clear()
random_01str(Nqubits_ZX,n)
print("Nqubits_ZX ,Z(0) or X(1) : {}".format(Nqubits_ZX))

for j in range(n):
    if not Nqubits_01[j] and (not Nqubits_ZX[j]):
        print("Alice generate |0> photon")
    elif Nqubits_01[j] and (not Nqubits_ZX[j]):
        print("Alice generate |1> photon")
    elif not Nqubits_01[j] and Nqubits_ZX[j]:
        print("Alice generate |+> photon")
    elif Nqubits_01[j] and Nqubits_ZX[j]:
        print("Alice generate |-> photon")
```

```python
# Message
Bob_Message_IY_circ0 = QuantumCircuit(q,c)  # Identity
Bob_Message_IY_circ0.iden(q[0])
#Bob_Message_IY_circ0.barrier(q)

Bob_Message_IY_circ1 = QuantumCircuit(q,c)  # Pauli-Y
Bob_Message_IY_circ1.y(q[0]) 
#Bob_Message_IY_circ1.barrier(q)


Bob_Message_H_circ0 = QuantumCircuit(q,c)  # no H
Bob_Message_H_circ0.barrier(q)

Bob_Message_H_circ1 = QuantumCircuit(q,c)  # H
Bob_Message_H_circ1.h(q[0]) 
Bob_Message_H_circ1.barrier(q)


# Decoy
Bob_Decoy_I_circ = QuantumCircuit(q,c)  # Identity
Bob_Decoy_I_circ.iden(q[0])
#Bob_Decoy_I_circ.barrier(q)

Bob_Decoy_Y_circ = QuantumCircuit(q,c)  # Pauli-Y
Bob_Decoy_Y_circ.y(q[0]) 
#Bob_Decoy_Y_circ.barrier(q)


Bob_Decoy_no_H_circ = QuantumCircuit(q,c)  # no H
Bob_Decoy_no_H_circ.barrier(q)

Bob_Decoy_H_circ = QuantumCircuit(q,c)  # H
Bob_Decoy_H_circ.h(q[0]) 
Bob_Decoy_H_circ.barrier(q)




#Bob_Message_IY_circ0.draw()
#Bob_Message_IY_circ1.draw()

#Bob_Message_H_circ0.draw()
#Bob_Message_H_circ1.draw()


#Bob_Decoy_I_circ.draw()
#Bob_Decoy_Y_circ.draw()
#Bob_Decoy_no_H_circ.draw()
#Bob_Decoy_H_circ.draw()
```

```python
# generate circuits
Bob_IY_message_circ = [Bob_Message_IY_circ0,Bob_Message_IY_circ1]
Bob_H_message_circ = [Bob_Message_H_circ0,Bob_Message_H_circ1]

Bob_Decoy_circ = [Bob_Decoy_I_circ,Bob_Decoy_Y_circ,Bob_Decoy_no_H_circ,Bob_Decoy_H_circ]

circuits_step4_2=[]
circuits_step4_2.clear()

for i in range(n):
    if not Bob_pre_key[i]: #decoy then m 
        #decoy
        if i%3==2: # Y
            if i%2==0: # H
                circuits_step4_2.append(circuits_step4_1[2*i]+Bob_Decoy_Y_circ+Bob_Decoy_H_circ)
            else: # no H
                circuits_step4_2.append(circuits_step4_1[2*i]+Bob_Decoy_Y_circ+Bob_Decoy_no_H_circ)
        else: # I
            if i%2==0: # H
                circuits_step4_2.append(circuits_step4_1[2*i]+Bob_Decoy_I_circ+Bob_Decoy_H_circ)
            else: # no H
                circuits_step4_2.append(circuits_step4_1[2*i]+Bob_Decoy_I_circ+Bob_Decoy_no_H_circ)
        #m
        circuits_step4_2.append(circuits_step4_1[2*i+1]+Bob_IY_message_circ[Bob_IY_message[i]]+Bob_H_message_circ[Bob_H_message[i]])
        
    else: # m then decoy
        #m
        circuits_step4_2.append(circuits_step4_1[2*i]+Bob_IY_message_circ[Bob_IY_message[i]]+Bob_H_message_circ[Bob_H_message[i]])
        #decoy
        if i%3==2: # Y
            if i%2==0: # H
                circuits_step4_2.append(circuits_step4_1[2*i+1]+Bob_Decoy_Y_circ+Bob_Decoy_H_circ)
            else: # no H
                circuits_step4_2.append(circuits_step4_1[2*i+1]+Bob_Decoy_Y_circ+Bob_Decoy_no_H_circ)
        else: # I
            if i%2==0: # H
                circuits_step4_2.append(circuits_step4_1[2*i+1]+Bob_Decoy_I_circ+Bob_Decoy_H_circ)
            else: # no H
                circuits_step4_2.append(circuits_step4_1[2*i+1]+Bob_Decoy_I_circ+Bob_Decoy_no_H_circ)

        
          
print("length of circuits_step4_2 : {}".format(len(circuits_step4_2)))
circuits_step4_2[0].draw()
#circuits_step4_2[1].draw()
```

```python
# Step6 : Send to Alice

# Alice knows how Bob operates decoy photons, so Alice can predict result of decoy photons.
print("decoy_01 ,0(0) or 1(1) : {}".format(decoy_01))
print("decoy_ZX ,Z(0) or X(1) : {}".format(decoy_ZX))
print("Bob operates Decoy --> I & Y : I I Y I I Y ... ; H : H () H () H () ...")

decoy_result_01=[]
decoy_result_ZX=[]
decoy_result_01.clear()
decoy_result_ZX.clear()
for i in range(n):
    if i%6==0 or i%6==4: # I H
        decoy_result_01.append(decoy_01[i])
        decoy_result_ZX.append(int(not decoy_ZX[i]))
    elif i%6==1 or i%6==3: # I
        decoy_result_01.append(decoy_01[i])
        decoy_result_ZX.append(decoy_ZX[i])
    elif i%6==2: # Y H
        decoy_result_01.append(int(not decoy_01[i]))
        decoy_result_ZX.append(int(not decoy_ZX[i]))
    elif i%6==5: # Y
        decoy_result_01.append(int(not decoy_01[i]))
        decoy_result_ZX.append(decoy_ZX[i])
        
print("decoy_result_01 ,0(0) or 1(1) : {}".format(decoy_result_01))
print("decoy_result_ZX ,Z(0) or X(1) : {}".format(decoy_result_ZX))
```

```python
# Alice gets circuits of decoy and message 

circuits_step6_decoy=[]
circuits_step6_message=[]

circuits_step6_decoy.clear()
circuits_step6_message.clear()

for i in range(n):
    if not Alice_pre_key[i]: # decoy then m
        circuits_step6_decoy.append(circuits_step4_2[2*i])
        circuits_step6_message.append(circuits_step4_2[2*i+1])
    else: # m then decoy
        circuits_step6_message.append(circuits_step4_2[2*i])
        circuits_step6_decoy.append(circuits_step4_2[2*i+1])
        

circuits_step6_decoy[0].draw()
#circuits_step6_message[0].draw()
```

```python
# Step8 : If no Eve, Alice recovers the state (N random qubits) to |0> by Y or H.

print("Nqubits_01 ,0(0) or 1(1) : {}".format(Nqubits_01))
print("Nqubits_ZX ,Z(0) or X(1) : {}".format(Nqubits_ZX))


Alice_recover_one_circ = QuantumCircuit(q,c)  # Y  #|1> --> |0>
Alice_recover_one_circ.y(q[0]) 
Alice_recover_one_circ.barrier(q)

Alice_recover_positive_circ = QuantumCircuit(q,c) # H  #|+> --> |0>
Alice_recover_positive_circ.h(q[0]) 
Alice_recover_positive_circ.barrier(q)

Alice_recover_negative_circ = QuantumCircuit(q,c) # Y H  #|-> --> |0>
Alice_recover_negative_circ.y(q[0])
Alice_recover_negative_circ.h(q[0]) 
Alice_recover_negative_circ.barrier(q)

#Alice_recover_one_circ.draw()
#Alice_recover_positive_circ.draw()
#Alice_recover_negative_circ.draw()


recover_state_circ=[]
recover_state_circ.clear()

for j in range(n):
    if not Nqubits_01[j] and (not Nqubits_ZX[j]):
        #print("|0> photon")
        recover_state_circ.append(circuits_step6_message[j])
    elif Nqubits_01[j] and (not Nqubits_ZX[j]):
        #print("|1> photon")
        recover_state_circ.append(circuits_step6_message[j]+Alice_recover_one_circ)
    elif not Nqubits_01[j] and Nqubits_ZX[j]:
        #print("|+> photon")
        recover_state_circ.append(circuits_step6_message[j]+Alice_recover_positive_circ)
    elif Nqubits_01[j] and Nqubits_ZX[j]:
        #print("|-> photon")
        recover_state_circ.append(circuits_step6_message[j]+Alice_recover_negative_circ)
        
        
recover_state_circ[0].draw()
```
